```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]
```

```python
results = {}
for element in queries:
    amount = len(element.split())
    if results.get(amount) is None:
        results[amount] = 1
    else:
        results[amount] +=1

print(results)
    
```

```python

for key, value in results.items():
    persentage = value / len(queries) *  100
    print(f'Поисковых запросов, содержащих {key} слов(а): {round(persentage, 2)}%')
```
